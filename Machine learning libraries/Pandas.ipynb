```python
data1 = [[122,"Kenny"],[151,"Sam"],[143,"James"],[123]]
df1 = pd.DataFrame(data1,columns=["Account no.","Name"])
df1
```

```python
data2 =[[122,19],[151,23],[143,21],[156,24]]
df2 = pd.DataFrame(data2,columns=["Account no.","Age"])
df2
```

#### DATAFRAME CONCATENATE

```python
df_list = pd.concat([df1,df2])
df_list
```

```python
df_list = pd.concat([df1,df2],axis=1)
df_list
```

#### DATAFRAME APPEND

```python
df1.append(df2)
```

####  DATAFRAME JOINS

```python
df1.set_index("Account no.").join(df2.set_index("Account no."),on= "Account no.",how = "inner")
```

```python
df1.join(df2,lsuffix="name",rsuffix="Age")
```

#### DATAFRAME MERGE

```python
pd.merge(df1,df2, on = "Account no.")
```

```python
pd.merge(df1,df2,on="Account no.",how="outer")
```

```python
pd.merge(df1,df2,on="Account no.",how="right")
```

```python
pd.merge(df1,df2,on="Account no.",how="left")
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

#### DATAFRAME RESHAPING

```python
data = [["kenny",19,"Male"],["Sam",22,"Male"]]
df3 = pd.DataFrame(data,columns=["Name","Age","Gender"])
df3
```

```python
df3_melt = df3.melt(id_vars=["Name"])
df3_melt
```

```python
df3_melt = df3.melt(id_vars=["Name"],value_vars=["Gender"])
df3_melt
```
