```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Automobile.csv")
df
```

### Scatterplot

```python
sns.relplot(x="body_style", y="price", data = df)
```

```python
sns.relplot(x="body_style", y="price", hue = "engine_size", data = df)
```

```python
sns.relplot(x="body_style", y="price", size = "engine_size", data = df);
```

### Jitter plot

```python
sns.catplot(x="body_style", y="price", data=df)
```

```python
sns.catplot(x="body_style", y="price", jitter = False,  data=df)
```

```python
sns.catplot(x="body_style", y="price", hue = "engine_size", data=df)
```

```python
sns.catplot(x="number_of_doors", y="price", kind = "swarm", data=df)
```

```python
sns.catplot(x="number_of_doors", y="price", hue = "engine_size", kind = "swarm", data=df)
```

### Box plot

```python
sns.catplot(x="number_of_doors", y="price", kind = "box", data=df)
```

```python
sns.catplot(x="fuel_type", y="price", hue = "engine_location", kind = "box", data=df)
```

### Violin plot

```python
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "violin", data=df)
```

```python
sns.catplot(x="fuel_type", y="price", hue = "aspiration", kind = "violin", split = True, data=df)
```

```python
sns.stripplot(x="fuel_type", y="price", data=df)
```

```python
sns.pairplot(df,diag_kind="kde")
```

### Bar plot

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="bar", data=df)
```

### Regression plot

```python
sns.regplot(x="city_mpg",y="price",data=df)
```

### Heatmap

```python
sns.heatmap(data=df.corr(),cmap="Blues")
```

```python
sns.heatmap(data=df.corr())
```

```python
sns.heatmap(data=df.corr(),annot=True)
```
